````python
# Write a function that returns the average of elements in a list.

# Constraints:

# Do NOT use sum()
# Handle empty list



def avg_list(lst):
    s=0
    for i in lst:
        s+=i
    return s/len(lst)

lst=[1,2,3,4,5,67,8]
avg_list(lst)
````

````python
# HW 2 — Remove Duplicates

# Write a function that removes duplicates from a list.

# Constraints:

# Do NOT use set()
# Maintain original order


def unique_list(lst):
    s=[]
    for i in lst:
        if i not in s:
            s.append(i)
    return s

lst=[1,2,3,44,44,5,67,8]
unique_list(lst)
````

````python
# HW 3 — Frequency Count

# Write a function that returns a dictionary with frequency of each element.

# Example:

# Input: [1,1,2,3,2]
# Output: {1:2, 2:2, 3:1}

def freq_count_list(lst):
    '''this function counts the no of frequency of elements and give a dict of it'''
    s={}
    for i in lst:
        if i not in s.keys():
            s[i]=1
        else :
            s[i]+=1
    return s

lst=[1,2,3,2,3,4,5,1,2,34,5,6,7,44,44,44,5,67,8]
freq_count_list(lst)
````

````python
# HW 4 — Extract Even Numbers

# Write a function that returns only even numbers from a list.

# Constraints:

# Return a new list (do not modify original)

def even_list(lst):
    new_lst=[]
    for i in lst:
        if i%2==0:
            new_lst.append(i)

    return new_lst

lst=[1,2,3,2,3,4,5,1,2,34,5,6,7,44,44,44,5,67,8]
even_list(lst)
````

````python
# HW 5 — Merge Two Lists into Tuple Pairs

# Write a function that takes two lists and returns a list of tuples.

# Example:

# Input: [1,2,3], [4,5,6]
# Output: [(1,4), (2,5), (3,6)]

def merge_list(a,b):
    output=[]
    for i in range(len(a)):
        c=[]
        c.append(a[i])
        c.append(b[i])   
        
        output.append(tuple(c))
    return output

a=[1,2,3]
b=[4,5,6]      

merge_list(a,b)
````

````python
 # PQ1: You have a function that takes *args. Write it so it returns the sum of all arguments. Now call it with 3 different combinations of inputs.
def sum_list(*args):
    s = 0
    for i in args:
        s+=i

    return s

sum_list(3,6,8)
````
