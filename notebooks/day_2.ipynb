## SQL
Topic: SELECT, WHERE, ORDER BY, LIMIT

### SQL1
Table employees(id, name, department, salary). Get all employees in Engineering with salary > 60000, ordered by salary DESC, top 5 only.

```python
sql_answer_1 = """
select * from employees
where salary > 6000
order by salary Desc
limit 5;
"""
print(sql_answer_1)
```

### SQL2
Find all employees whose name starts with 'A' OR who earn more than 90000. Use WHERE with LIKE and OR.

```python
sql_answer_2 = """
select name from employees
where lower(name) LIKE "a%" OR salary > 90000
"""
print(sql_answer_2)
```
