## 📘 Concepts to Know Before You Start

### File Operations
- `open(file, mode)` — opens a file. Modes: `'r'` read, `'w'` write (overwrites), `'a'` append, `'rb'` read binary
- Always use `with open(...) as f:` — it auto-closes the file even if an error happens
- `f.read()` reads entire file, `f.readlines()` reads line by line into a list, `f.write()` writes string

### File Paths
- `os.path.join('folder', 'file.txt')` → works on ALL operating systems (Windows uses `\`, Mac/Linux use `/`)
- `pathlib.Path` is the modern way — `Path('folder') / 'file.txt'` does the same thing but cleaner
- `Path.exists()`, `Path.stem` (filename without extension), `Path.parent` (parent folder), `Path.rglob('*.csv')` (find all csv files recursively)

### Exception Handling
- `try` → code that might fail
- `except SpecificError` → catches only that error (preferred over catching everything)
- `else` → runs only if NO exception occurred
- `finally` → ALWAYS runs, error or not (use for cleanup like closing files)
- Always catch specific exceptions (`FileNotFoundError`, `ValueError`) not just `Exception` — otherwise you hide bugs

### Why this matters in Big Data
In data pipelines, files are constantly being read, written, and moved. A pipeline that crashes without proper error handling can corrupt data or leave partial writes. Every real DE writes defensive file code.

### 📘 Quick concept reminder:
- `GROUP BY` groups rows with the same value together
- Aggregate functions work on groups: `COUNT()`, `SUM()`, `AVG()`, `MAX()`, `MIN()`
- `WHERE` filters BEFORE grouping (filters individual rows)
- `HAVING` filters AFTER grouping (filters groups)
- Rule: if you use an aggregate in your condition, use HAVING not WHERE
